```python
df = spark.read.csv(path="/Volumes/quickstart_catalog/quickstart_schema/sandbox/dataset/employee.csv", 
                    header=True,
                    sep ="|",
                    inferSchema = True)
```

```python
display(df)
```

```python
df.na.drop().show()
```

```python
from pyspark.sql.functions import col
df.filter((col("id").isNotNull()) & (col("name").isNotNull())).display()
 
```

```python
df.na.fill(0).na.fill("NULL in source").display()
```

```python
df.na.fill({
    "id":-1,
    "exp":0,
    "gen":"NA"
}).display()
```

```python
from pyspark.sql.functions import avg
mean = df.select(avg("exp")).collect()[0][0]
df.na.fill({
    "id":-1,
    "exp":int(mean),
    "gen":"NA"
}).display()
```
